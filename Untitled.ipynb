```python
import nltk
nltk.download('gutenberg')
import pandas 
import numpy
from nltk.text import Text
alice = nltk.corpus.gutenberg.raw('carroll-alice.txt')
from nltk import PorterStemmer
from nltk import LancasterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
# split into words by white space
words = alice.split()
words = [word.lower() for word in words]
print(words[:100])
```

```python
import string
table = str.maketrans('', '', string.punctuation)
words = [w.translate(table) for w in words]
words
```

```python
porter = PorterStemmer()
for w in words:
   print(w, " : ", porter.stem(w))
```

```python
lancaster= LancasterStemmer()
for w in words:
   print(w, " : ", lancaster.stem(w))
```

```python
# I think the first difference is that Lancaster is more aggresive rooting than Porter, 
#Lancaster took more time process than Proter in my kernel
# Proter produced more better favorable results than the other
```

```python
nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()
for w in words:
   print(w, " : ", wordnet_lemmatizer.lemmatize(w))
```

```python
#Advantages of Stemming is of course the shortening of word for better indexing and they are fast to process.
#Stemming can't relate words which have different forms based on grammatical constructs.
#stemming can't prune them to the common form.
```
